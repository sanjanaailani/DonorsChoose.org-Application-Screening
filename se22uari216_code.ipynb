```python
import numpy as np
import pandas as pd
import datetime
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('train.csv', parse_dates=['project_submitted_datetime'])
test = pd.read_csv('test.csv', parse_dates=['project_submitted_datetime'])
print(train.shape)
print(test.shape)
```

```python
train.drop_duplicates()
total = train.isnull().sum().sort_values(ascending = False)
missing_train_data  = pd.concat([total], axis=1, keys=['Total'])
missing_train_data.head()
```

```python
test.drop_duplicates()
total = test.isnull().sum().sort_values(ascending = False)
missing_test_data  = pd.concat([total], axis=1, keys=['Total'])
missing_test_data.head()
```

```python
train.groupby('school_state').agg({'project_is_approved': ['mean', 'count']}).reset_index().sort_values([('project_is_approved', 'mean')], ascending=False).reset_index(drop=True)
```

```python
train['project_submitted_datetime'] = pd.to_datetime(train['project_submitted_datetime'])
test['project_submitted_datetime'] = pd.to_datetime(test['project_submitted_datetime'])

train["year"] = train["project_submitted_datetime"].dt.year
train["month"] = train["project_submitted_datetime"].dt.month
train['weekday'] = train['project_submitted_datetime'].dt.weekday
train["hour"] = train["project_submitted_datetime"].dt.hour
train["month_Day"] = train['project_submitted_datetime'].dt.day
train["year_Day"] = train['project_submitted_datetime'].dt.dayofyear
train = train.drop('project_submitted_datetime', axis=1)

test["year"] = test["project_submitted_datetime"].dt.year
test["month"] = test["project_submitted_datetime"].dt.month
test['weekday'] = test['project_submitted_datetime'].dt.weekday
test["hour"] = test["project_submitted_datetime"].dt.hour
test["month_Day"] = test['project_submitted_datetime'].dt.day
test["year_Day"] = test['project_submitted_datetime'].dt.dayofyear
test = test.drop('project_submitted_datetime', axis=1)
```

```python
r = ['project_title','project_essay_1','project_essay_2','project_essay_3','project_essay_4','project_resource_summary']

train_data = train.drop(r, axis = 1)
test_data = test.drop(r, axis = 1)
```

```python
train_data.to_csv("train_new.csv", index=False)
```

```python
test_data.to_csv("test_new.csv", index=False)
```

```python
def encode_features(df_train, df_test):
    features = [
    'teacher_id',
    'teacher_prefix',
    'school_state',
    'project_grade_category',
    'project_subject_categories',
    'project_subject_subcategories',
    ]
    df_combined = pd.concat([df_train[features], df_test[features]])

    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(df_combined[feature].astype(str))
        df_train[feature] = le.transform(df_train[feature].astype(str))
        df_test[feature] = le.transform(df_test[feature].astype(str))
    return df_train, df_test

train, test = encode_features(train_data, test_data)
```

```python
X = train_data.drop(columns=['project_is_approved'])
y=train_data['project_is_approved']
X_train_rf, X_valid_rf, y_train_rf, y_valid_rf = train_test_split(X, y, test_size=0.20, random_state=4018)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train_rf, y_train_rf)

rf_pred_valid = model_rf.predict(X_valid_rf)

auc_rf = roc_auc_score(y_valid_rf, rf_pred_valid)
precision_rf = precision_score(y_valid_rf, rf_pred_valid)
recall_rf = recall_score(y_valid_rf, rf_pred_valid)
f1_rf = f1_score(y_valid_rf, rf_pred_valid)
accuracy_rf = accuracy_score(y_valid_rf, rf_pred_valid)

print('AUC (Random Forest):', auc_rf)
print('Precision (Random Forest):', precision_rf)
print('Recall (Random Forest):', recall_rf)
print('F1-score (Random Forest):', f1_rf)
print('Accuracy (Random Forest):', accuracy_rf)
```

```python
rf_predictions = model_rf.predict(test_data)
test_data['predicted_label'] = rf_predictions
test_data.to_csv("predict_se22uari216.csv", index=False)
```
